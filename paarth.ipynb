```python
from langchain.agents.agent_types import AgentType
from langchain_community.chat_models import ChatOpenAI
from langchain_community.llms import OpenAI
from langchain_experimental.agents.agent_toolkits import create_csv_agent
import os
from dotenv import load_dotenv
load_dotenv()
import pandas as pd
```

```python
df=pd.read_csv("paarth.csv")
```

```python
curr_folder=os.getcwd()
file_path=os.path.join(curr_folder,"paarth.csv")
file_path
```

```python
agent = create_csv_agent(OpenAI(temperature=0, max_tokens=500), file_path, verbose=True)
prompt="Assume that you are a very good data analyst. You are given the stock data of two firms ADANIPORTS and AXISBANK shown in the Symbol column. \
        The format of the date column is YYYY-MM-DD. VWAP in the dataframe is the volume-weighted average price of the \
        stock. Note: In the following prompts, don't try to do mathematical calculation on the columns Date and Symbol. \
        Note: Date and Symbol are not numeric, so any mathematical function should not be applied on these columns.\
        \
        Task #1: Select the rows of AXISBANK where the deliverable volume is greater than 4 million. ONLY RETURN THE INDEX OF THE ROWS AS OUTPUT, DON'T RETURN ANYTHING ELSE."
response=agent.run(prompt)
print(response)
```

```python
indices = [int(idx) for idx in response.split("[")[1].split("]")[0].split(",")]
```

```python
df.loc[indices]
```
